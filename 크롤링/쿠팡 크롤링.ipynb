```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import pandas as pd
import os
```

```python
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.coupang.com/vp/products/5080540471?itemId=6891539307&vendorItemId=4821810379&q=%EB%B0%A5%EC%86%A5&itemsCount=36&searchId=e9ced183180f4dd6aca6452d60d59762&rank=0&isAddedCart=')
time.sleep(5)
```

```python
a=driver.find_element(By.XPATH, "//*[@id='btfTab']/ul[1]/li[2]")
a.click()
# b=driver.find_element(By.XPATH, '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/div[3]/button[2]')
```

```python
i = 1
j = 2
k= 1
box_list=[]
while True:
    if j > 2 and i > 5:
        driver.find_element(By.XPATH, '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/div[3]/button['+str(j)+']').click()
        time.sleep(2)
        i = 1
   
    elif j == 12:
        j = 2
        k += 1
        i = 1
    else:        
        boxes=a.find_elements(By.XPATH,'/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article['+str(i)+']')
        for box in boxes:
            ID=box.find_element(By.XPATH, '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article['+str(i)+']/div[1]/div[2]/span').text
            date=box.find_element(By.XPATH, '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article['+str(i)+']/div[1]/div[3]/div[2]').text
            product=box.find_element(By.XPATH, '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article['+str(i)+']/div[1]/div[5]').text
            title=box.find_element(By.XPATH, '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article['+str(i)+']/div[3]').text
            comment=box.find_element(By.XPATH, '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article['+str(i)+']/div[4]/div').text
            box_item={'ID' : ID, 'date':date, 'product' : product, 'title' : title,'comment' : comment}
            box_list.append(box_item)
            i += 1
        if i > 5:
            j += 1
    if i == 4:
        break
print(box_list)
```
